```python
# Import required libraries
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("Telco customer churn.csv")

# Check shape
print(f"Dataset shape: {df.shape}")

# Preview data
display(df.head())

# Check basic information
df.info()

# Check number of missing values
missing_values = df.isnull().sum()
print("Missing values:\n", missing_values[missing_values > 0])

# Check unique values per column
unique_counts = df.nunique().sort_values()
print("Unique values per column:\n", unique_counts)
```

```python
# Convert TotalCharges to numeric, coerce invalid entries to NaN
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Check if any NaN appeared after conversion
print("Missing after conversion:\n", df.isnull().sum())

# Drop rows with missing TotalCharges (if any)
df = df[df['TotalCharges'].notnull()].copy()

# Reset index after drop
df.reset_index(drop=True, inplace=True)

# Cast SeniorCitizen to object (categorical)
df['SeniorCitizen'] = df['SeniorCitizen'].astype(str)
```

```python
# Separate target variable
target = 'Churn'

# Identify numerical features
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
numerical_features.remove('tenure') if 'tenure' in numerical_features else None  # treat as ordered categorical if needed

# Identify categorical features
categorical_features = df.select_dtypes(include='object').columns.tolist()
categorical_features.remove(target)

# Print summary
print("Numerical features:", numerical_features)
print("Categorical features:", categorical_features)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot churn distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Churn', palette='pastel')
plt.title('Churn Distribution')
plt.xlabel('Churn')
plt.ylabel('Count')
plt.show()
```

```python
# Contract type vs churn
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='Contract', hue='Churn', palette='Set2')
plt.title('Churn by Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Count')
plt.legend(title='Churn')
plt.show()
```

```python
# MonthlyCharges vs churn (boxplot)
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Churn', y='MonthlyCharges', palette='coolwarm')
plt.title('Monthly Charges by Churn Status')
plt.show()
```

```python
# Tenure distribution by churn
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='tenure', hue='Churn', multiple='stack', bins=30, palette='Set1')
plt.title('Tenure Distribution by Churn')
plt.xlabel('Tenure (Months)')
plt.ylabel('Count')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(8, 6))
corr = df[['tenure', 'MonthlyCharges', 'TotalCharges']].corr()
sns.heatmap(corr, annot=True, cmap='Blues', fmt=".2f")
plt.title('Correlation Between Numeric Features')
plt.show()
```

```python
# Encode target variable: Churn â†’ 1 for 'Yes', 0 for 'No'
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
```
